```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img=cv2.imread("/home/jesterrexx/Pictures/Cars/469439.jpg")

print('This image is: ',type(img),"with dimensions",img.shape)
```

```python
%matplotlib inline
img_copy=np.copy(img)
img_copy=cv2.cvtColor(img_copy,cv2.COLOR_BGR2RGB)
plt.imshow(img_copy)
```

```python
# Define
lower_=np.array([230,200,0])
upper_=np.array([50,50,255])
```

```python
# Define the masked Area
mask=cv2.inRange(img_copy,lower_,upper_)
plt.imshow(mask,cmap='gray')
```

```python
masked_img=np.copy(img_copy)
masked_img[mask!=0]=[0,0,0]
plt.imshow(masked_img)
```

# Mask and add a background image

```python
background_img=cv2.imread("/home/jesterrexx/Pictures/New York/andrae-ricketts-0_la3_Slfwk-unsplash.jpg")
background_img=cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)

crop_background=background_img[0:2160,0:3840]

crop_background[mask==0]=[0,0,0]
```

```python
plt.imshow(crop_background)
```

```python
complete_image=crop_background+masked_img

plt.imshow(complete_image)
```
